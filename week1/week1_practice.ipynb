## 1. Creating Arrays

NumPy’s core object is the **ndarray**. Every array has two key properties you should always check first:
- `.shape` — the size along each dimension
- `.dtype` — the data type of each element

````python
# From a Python list
a = np.array([1, 2, 3, 4, 5])
print("a       =", a)
print("a.shape =", a.shape)      # (5,)  — 1-D, 5 elements
print("a.dtype =", a.dtype)      # int64
````

````python
# 2-D array (matrix)
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print("b.shape =", b.shape)         # (2, 3) — 2 rows, 3 columns
print("b.dtype =", b.dtype)
print(f"b =\n{b}")
````

````python
# Filled arrays
print(f"zeros:\n{np.zeros((2, 3))}")       # all 0s  (float64 by default)
print(f"ones:\n{np.ones((2, 3))}")          # all 1s
````

````python
# Match an existing array's shape — useful when you need a blank canvas the same size
ref = np.array([[1, 2], [3, 4]])
print(f"zeros_like:\n{np.zeros_like(ref)}")   # same shape, all 0
print(f"ones_like:\n{np.ones_like(ref)}")     # same shape, all 1
print(f"copy:\n{ref.copy()}")                  # independent deep copy
````

## 2. arange & linspace

`np.arange` works like Python's `range()` but returns a NumPy array.
`np.linspace` gives you exactly *N* evenly spaced values between two endpoints.

Both are essential for generating pixel coordinates and value ranges.

````python
# arange — like Python's range(), but returns a NumPy array
print("arange(5)       =", np.arange(5))            # [0, 1, 2, 3, 4]  — stop only
print("arange(2, 7)    =", np.arange(2, 7))          # [2, 3, 4, 5, 6]  — start, stop
print("arange(0, 10, 2)=", np.arange(0, 10, 2))      # [0, 2, 4, 6, 8]  — start, stop, step
print("arange(0, 256)  → length", len(np.arange(0, 256)))  # 256 values: the full pixel range!
````

````python
# linspace — exactly N values, evenly spaced (endpoints included)
print("linspace(0, 1, 5)  =", np.linspace(0, 1, 5))     # [0, 0.25, 0.5, 0.75, 1.0]
print("linspace(0, 255, 4)=", np.linspace(0, 255, 4))    # [0, 85, 170, 255]
````

## 3. Indexing & Slicing

Works like Python lists, but extended to multiple dimensions: `array[row, col]`.

These same operations will let you **crop**, **flip**, and **downsample** images.

````python
m = np.arange(12).reshape(3, 4)   # 3×4 matrix  [0..11]
print(f"m =\n{m}")
````

````python
print("m[0, 2]      =", m[0, 2])        # single element: row 0, col 2
print("m[1]         =", m[1])            # entire row 1
print("m[:, 0]      =", m[:, 0])         # entire column 0  (: means 'all rows')
print(f"m[0:2, 1:3]  =\n{m[0:2, 1:3]}")  # sub-matrix: rows 0-1, cols 1-2
````

````python
# Negative indexing & reversing
print("last row      =", m[-1])           # last row
print(f"rows reversed =\n{m[::-1]}")       # reverse row order → vertical flip
print(f"cols reversed =\n{m[:, ::-1]}")    # reverse column order → horizontal flip

# Step slicing: [start:stop:step] — take every Nth element
print("\nevery 2nd col =", m[0, ::2])      # step=2 → cols 0, 2
print(f"every 2nd row =\n{m[::2]}")        # rows 0, 2
print(f"every 2nd both =\n{m[::2, ::2]}")  # subsample rows AND cols by 2
````

## 4. Element-wise Arithmetic

Operators `+`, `-`, `*`, `/`, `**` work **element-by-element**. A scalar is automatically **broadcast** to match the array shape.

This is how we'll brighten, darken, and adjust contrast of images — just add, subtract, or multiply pixel values.

````python
# Two arrays of the same shape → element-wise
v = np.array([10, 20, 30])
w = np.array([1, 2, 3])
print("v + w =", v + w)             # [11, 22, 33]
print("v * w =", v * w)             # [10, 40, 90]
````

````python
# np.clip — critical for image processing!
# Pixel values must stay in [0, 255]. Clip enforces this.
x = np.array([-10, 50, 200, 300])
print("before clip:", x)
print("after  clip:", np.clip(x, 0, 255))   # [-10→0, 300→255]
````

## 5. Reshape & Dimensions

`reshape` changes the **logical shape** without changing data. We’ll use this to flatten images into vectors (for ML) or add a channel dimension to grayscale images.

````python
a = np.arange(12)
print("1-D      =", a)              # [0, 1, ..., 11]   shape (12,)

b = a.reshape(3, 4)                  # → shape (3, 4)
print(f"reshaped =\n{b}")

c = b.reshape(6, 2)                  # → shape (6, 2)
print(f"reshaped again =\n{c}")

d = b.reshape(-1)                    # -1 = "infer this dimension" → back to 1-D
print("back to 1-D =", d)
````

````python
a = np.zeros((4, 5))               # shape (4, 5)
print("original shape   :", a.shape)

# expand_dims — add a dimension
b = np.expand_dims(a, axis=2)       # (4, 5) → (4, 5, 1)
print("expand axis=2    :", b.shape)

c = np.expand_dims(a, axis=0)       # (4, 5) → (1, 4, 5)
print("expand axis=0    :", c.shape)

# Shorthand: index with None (same as np.newaxis)
d = a[:, :, None]                   # (4, 5) → (4, 5, 1)
print("[:, :, None]     :", d.shape)

e = a[None, :, :]                   # (4, 5) → (1, 4, 5)
print("[None, :, :]     :", e.shape)
````

````python
# squeeze — remove all size-1 dimensions
f = np.zeros((1, 4, 5, 1))
print("before squeeze:", f.shape)    # (1, 4, 5, 1)
print("after  squeeze:", np.squeeze(f).shape)  # (4, 5)
````

## 6. Concatenate & Stack

- `np.concatenate` joins arrays along an **existing** axis.
- `np.stack` joins arrays along a **new** axis.

The key use-case: `np.stack([r, g, b], axis=2)` builds an RGB image from three channels.

````python
p = np.ones((2, 3))
q = np.zeros((2, 3))

# concatenate along axis 0 (stack rows)
r = np.concatenate([p, q], axis=0)
print("concat axis=0 :", r.shape)   # (4, 3)
print(r)
print()

# concatenate along axis 1 (append columns)
s = np.concatenate([p, q], axis=1)
print("concat axis=1 :", s.shape)   # (2, 6)
print(s)
````

````python
# stack — adds a NEW axis
t = np.stack([p, q], axis=0)
print("stack axis=0 :", t.shape)    # (2, 2, 3) — two (2,3) arrays into a new dim

u = np.stack([p, q], axis=2)
print("stack axis=2 :", u.shape)    # (2, 3, 2)
````
